# Load and plot data

```python
import pathlib
OUT_PATH = pathlib.Path("output")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
run_id = "default"
param = 10
```

```python
# Set Output-Folders
file_path = OUT_PATH / run_id / str(param)
file_path.mkdir(parents=True, exist_ok=True)
```

```python
size = 100
```

```python
data = pd.read_pickle(file_path / "data.pkl")
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for label, df in data.groupby('col2'):
    df.col1.plot(kind="kde", ax=ax, label=label)
plt.legend();
```

```python
fig.savefig(file_path / "result.pdf", bbox_inches='tight')
```
